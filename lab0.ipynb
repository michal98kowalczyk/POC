```
import cv2
from matplotlib import pyplot as plt
import os
import numpy as np


if not os.path.exists("mandril.jpg"):
    !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/01_Intro/mandril.jpg

I = cv2.imread('mandril.jpg')
I =cv2.cvtColor(I,cv2.COLOR_RGB2GRAY)
plt.imshow(I)
plt.xticks([]),plt.yticks([])
plt.gray()
plt.show()
#zapisanie obrazu
cv2.imwrite("saved.png",I)

#siatka
xx, yy = np.mgrid[0:I.shape[0], 0:I.shape[1]]
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(xx, yy, I ,rstride=1, cstride=1, cmap=plt.get_cmap('gray'),linewidth=0)
plt.show()
S = I[10,:]
K=I[:,10]
f,(ax1,ax2)=plt.subplots(1,2)
ax1.set_title("Pierwszy")
ax2.set_title("Drugi")
ax1.plot(S)
ax2.plot(K)
plt.show()
#mapy kolor√≥w
plt.imshow(I,cmap=plt.get_cmap('Oranges'))
plt.show()

plt.imshow(I,cmap=plt.get_cmap('Blues'))
plt.show()

plt.imshow(I,cmap=plt.get_cmap('YlOrBr'))
plt.show()

plt.imshow(I,cmap=plt.get_cmap('BuGn'))
plt.show()
```
