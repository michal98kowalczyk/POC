```
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math
import os

# Load required files
if not os.path.exists("dwieFale.bmp") :
    !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/dwieFale.bmp 
if not os.path.exists("kolo.bmp") :
    !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/kolo.bmp 
if not os.path.exists("kwadrat.bmp") :
    !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/kwadrat.bmp 
if not os.path.exists("kwadrat45.bmp") :
    !wget  --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/kwadrat45.bmp 
if not os.path.exists("kwadratKL.bmp") :
    !wget  --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/kwadratKL.bmp 
if not os.path.exists("kwadratS.bmp") :
    !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/kwadratS.bmp 
if not os.path.exists("kwadratT.bmp") :
    !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/kwadratT.bmp 
if not os.path.exists("lena.bmp") :
    !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/lena.bmp 
if not os.path.exists("trojkat.bmp") :
    !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/trojkat.bmp 

# I_Fale = cv2.imread('dwieFale.bmp', cv2.IMREAD_GRAYSCALE)

# figFale, axsFale = plt.subplots()

# axsFale.imshow(I_Fale, 'gray', vmin=0, vmax=256)
# axsFale.axis('off')

# figFale.show()
def showImages(img,mag, ang,Alog):
  fig, axs = plt.subplots(1,4)
  axs[0].imshow(img, 'gray')
  axs[0].set_title('img')
  axs[0].axis('off')
  axs[1].imshow(mag, 'gray')
  axs[1].axis('off')
  axs[1].set_title('mag')
  axs[2].imshow(ang, 'gray')
  axs[2].axis('off')
  axs[2].set_title('ang')
  axs[3].imshow(Alog, 'gray')
  axs[3].axis('off')
  axs[3].set_title('Alog')
  plt.show()
  
 

def fourier(img):
  dft = cv2.dft(np.float32(img),flags=cv2.DFT_COMPLEX_OUTPUT)
  gx = np.fft.fftshift(dft[:, :, 0])
  gy = np.fft.fftshift(dft[:, :, 1])


  
  mag, ang = cv2.cartToPolar(gx,gy)
  Alog = np.log10(mag + 1)


 

  return mag, ang,Alog
 


def fRC(img):
  FRow = np.fft.fft(img, axis=0)
  FRes = np.fft.fft(FRow, axis=1)
  m1 = cv2.magnitude(np.real(FRow),np.imag(FRow))
  m2 = cv2.magnitude(np.real(FRes),np.imag(FRes))
  # m2 = cv2.magnitude(FRes[:,:,0],FRes[:,:,1])
  # img = abs(m1-m2)
  # img = cv2.absdiff(m1,m2)
  img= abs(m1-m2)
  plt.imshow(img,'gray',vmin=0 , vmax=256)
  plt.title('Compare')
  plt.show()


# ---zad1
fale = cv2.imread('dwieFale.bmp', cv2.IMREAD_GRAYSCALE)
mag, ang,Alog = fourier(fale)
showImages(fale,mag, ang,Alog)
fRC(fale)


kolo = cv2.imread('kolo.bmp', cv2.IMREAD_GRAYSCALE)
mag, ang,Alog = fourier(kolo)
showImages(kolo,mag, ang,Alog)
fRC(kolo)

kwadrat = cv2.imread('kwadrat.bmp', cv2.IMREAD_GRAYSCALE)
mag, ang,Alog = fourier(kwadrat)
showImages(kwadrat,mag, ang,Alog)
fRC(kwadrat)

trojkat = cv2.imread('trojkat.bmp', cv2.IMREAD_GRAYSCALE)
mag, ang,Alog = fourier(trojkat)
showImages(trojkat,mag, ang,Alog)
fRC(trojkat)





# # ----zad2
kwadratT = cv2.imread('kwadratT.bmp', cv2.IMREAD_GRAYSCALE)
kwadrat45 = cv2.imread('kwadrat45.bmp', cv2.IMREAD_GRAYSCALE)
kwadratS = cv2.imread('kwadratS.bmp', cv2.IMREAD_GRAYSCALE)
kwadratKL = cv2.imread('kwadratKL.bmp', cv2.IMREAD_GRAYSCALE)


  

# translacja
def _translation(img):


  rows,cols = img.shape
  M = np.float32([[1,0,100],[0,1,50]])
  dst = cv2.warpAffine(img,M,(cols,rows))

  mag, ang,Alog = fourier(img)
  showImages(img,mag, ang,Alog)
  mag, ang,Alog = fourier(dst)
  showImages(dst,mag, ang,Alog)


_translation(kwadrat)
_translation(kwadratT)


def _rotation(img):
  rows,cols = img.shape

  M = cv2.getRotationMatrix2D((cols/2,rows/2),60,1)
  dst = cv2.warpAffine(img,M,(cols,rows))
  mag, ang,Alog = fourier(img)
  showImages(img,mag, ang,Alog)
  mag, ang,Alog = fourier(dst)
  showImages(dst,mag, ang,Alog)

_rotation(kwadrat)
_rotation(kwadrat45)


def _scaling(img):
  res = cv2.resize(img,None,fx=0.5, fy=1, interpolation = cv2.INTER_CUBIC)

  mag, ang,Alog = fourier(img)
  showImages(img,mag, ang,Alog)
  mag, ang,Alog = fourier(res)
  showImages(res,mag, ang,Alog)


_scaling(kwadrat)
_scaling(kwadratS)


def linComb(img1,img2):
  h1,w1 = img1.shape
  h2,w2 = img2.shape

  img2 =cv2.resize(img2,(w1,h1))
  res =cv2.addWeighted(img1,0.2,img2,0.8,0)

  
  mag, ang,Alog = fourier(res)
  showImages(res,mag, ang,Alog)

print("kwadrat + kwadrat45")
linComb(kwadrat,kwadrat45)
print("kwadrat + kwadratKL")
linComb(kwadrat,kwadratKL)
print("kwadratKL + kwadrat45")
linComb(kwadratKL,kwadrat45)







# ---zad3 

kolo = cv2.imread('kolo.bmp', cv2.IMREAD_GRAYSCALE)

dft = cv2.dft(np.float32(kolo),flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
rows, cols = kolo.shape
crow,ccol = rows//2 , cols//2
mask = np.zeros((rows,cols,2),np.uint8)
mask[crow-30:crow+30, ccol-30:ccol+30] = 1
fshift = dft_shift*mask
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])

fig, axs = plt.subplots(1,2)
axs[0].imshow(kolo, 'gray')
axs[0].set_title('przed')
axs[0].axis('off')
axs[1].imshow(img_back, 'gray')
axs[1].axis('off')
axs[1].set_title('po')
plt.show()






# zad4
lena = cv2.imread('lena.bmp', cv2.IMREAD_GRAYSCALE)
mag, ang,Alog = fourier(lena)
showImages(lena,mag, ang,Alog)



def _filter(img,FilterF):

  imgSize = img.shape
  FSpaceRows = 2 * np.fft.fftshift(np.fft.fftfreq(imgSize[0]))
  FSpaceRowsM = np.outer(FSpaceRows, np.ones([1, imgSize[1]]))
  FSpaceCols = 2 * np.fft.fftshift(np.fft.fftfreq(imgSize[1]))
  FSpaceColsM = np.outer(np.ones([1, imgSize[0]]), FSpaceCols)

  FreqR = np.sqrt(np.square(FSpaceRowsM) + np.square(FSpaceColsM))



  dft = cv2.dft(np.float32(img), flags = cv2.DFT_COMPLEX_OUTPUT)
  dft_shift = np.fft.fftshift(dft)
  FilterF3 = np.repeat(FilterF[:, :, np.newaxis], 2, axis=2)

  dst_shiftF = dft_shift * FilterF3
  dst_ishiftF = np.fft.ifftshift(dst_shiftF)


  figFilter = plt.figure()
  axsFilter = figFilter.add_subplot(projection='3d')
  axsFilter.plot_surface(FSpaceRowsM, FSpaceColsM, FilterF, rstride=1, cstride=1, cmap=plt.get_cmap('gray'), linewidth=0)
  figFilter.show()

_filter(lena,FreqR <= 0.1)
_filter(lena,(FreqR >= 0.1) & (FreqR <= 0.3))
_filter(lena,FreqR >= 0.3)





# --zad 5
if not os.path.exists("literki.bmp") :
    !wget https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/literki.bmp --no-check-certificate
if not os.path.exists("wzorA.bmp") :
    !wget https://raw.githubusercontent.com/vision-agh/poc_sw/master/08_Fourier/wzorA.bmp --no-check-certificate


literki = cv2.imread('literki.bmp',cv2.IMREAD_GRAYSCALE)
wzor = cv2.imread('wzorA.bmp',cv2.IMREAD_GRAYSCALE)

literkiF = cv2.dft(np.float32(literki), flags = cv2.DFT_COMPLEX_OUTPUT)

wzorRot = np.rot90(wzor,2)



wzorH,wzorW = wzor.shape
literkiH, literkiW = literki.shape
img = cv2.copyMakeBorder(wzorRot,0,literkiH-wzorH,0,literkiW-wzorW,cv2.BORDER_CONSTANT,value=0)
dft = cv2.dft(np.float32(img), flags = cv2.DFT_COMPLEX_OUTPUT)
res = (np.fft.fftshift(dft[:, :, 0]) + np.fft.fftshift(dft[:, :, 1]) * 1j) * (np.fft.fftshift(literkiF[:, :, 0]) + np.fft.fftshift(literkiF[:, :, 1]) * 1j)

CompMat = cv2.merge([np.real(res), np.imag(res)])
idft = abs(cv2.idft(CompMat, flags=cv2.DFT_COMPLEX_INPUT))
res=cv2.morphologyEx(idft, cv2.MORPH_TOPHAT, np.ones((3, 3), np.uint8))
mag = cv2.magnitude(res[:,:,0], res[:,:,1])

fig, axs = plt.subplots(1, 3)
axs[0].imshow(literki, cmap = 'gray')
axs[0].set_title('literki')
axs[1].imshow(wzor, cmap = 'gray')
axs[1].set_title('wzorA')
axs[2].imshow(mag, cmap = 'gray')
axs[2].set_title('mag')
```
