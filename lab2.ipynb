```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

if not os.path.exists("lut.py"):
    !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/lut.py

import lut

plt.plot(lut.kwadratowa)
plt.show()

# pobranie obrazkow
if not os.path.exists("lena.bmp"):
  !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/lena.bmp

if not os.path.exists("jet.bmp"):
   !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/jet.bmp

lena = cv2.imread('lena.bmp')
plt.imshow(lena)
plt.show()
lena2 = cv2.LUT(lena,lut.log)
plt.imshow(lena2)
plt.show()


def LUT(img,tab):
  img2 = cv2.LUT(img,tab)
  f, (ax1, ax2, ax3) = plt.subplots(1,3)
  ax1.set_title('Before')
  ax2.set_title('Table')
  ax2.set_aspect('equal')
  ax3.set_title("After")
  ax1.imshow(img)
  ax2.plot(tab)
  ax3.imshow(img2)
  plt.show()

LUT(lena,lut.log)
LUT(lena,lut.pila)
LUT(lena,lut.odwlog)
LUT(lena,lut.wykladnicza)
LUT(lena,lut.kwadratowa)
LUT(lena,lut.odwrotna)
LUT(lena,lut.pierwiastkowa)


plt.imshow(lena)
plt.show()
# dodawania
jet = cv2.imread('jet.bmp')
plt.imshow(jet)
plt.show()
add = cv2.add(lena,jet)
plt.imshow(add)
plt.show()


add2 = cv2.add(lena.astype('uint16'),jet2.astype('uint16'))
plt.imshow((add2/4).astype('uint8'))
plt.show()


# kombinacja liniowa
line = cv2.addWeighted(lena,0.7,jet,0.3,100)
plt.imshow(line)
plt.show()


# odejmowanie
sub = cv2.subtract(lena,jet)
plt.imshow(sub)
plt.show()

sub2 = cv2.subtract(lena.astype('int16'),jet2.astype('int16'))
sub2 = np.abs(sub2)
plt.imshow(sub2)
plt.show()


# mnozenie 
mult = cv2.multiply(lena,jet)
plt.imshow(mult)
plt.show()

if not os.path.exists("kolo.bmp"):
   !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/kolo.bmp

kolo = cv2.imread('kolo.bmp')
mult2 = cv2.multiply(lena,kolo.astype('bool').astype('uint8'))
plt.imshow(mult2)
plt.show()


# negatyw
neg = cv2.bitwise_not(lena)
plt.imshow(neg)
plt.show()

#operacje logiczne

if not os.path.exists("kwadrat.bmp"):
   !wget --no-check-certificate https://raw.githubusercontent.com/vision-agh/poc_sw/master/02_Point/kwadrat.bmp

kwadrat = cv2.imread('kwadrat.bmp')
plt.imshow(kolo)
plt.show()
plt.imshow(kwadrat)
plt.show()


plt.imshow( kolo )
plt.show()
plt.imshow( ~kolo )
plt.show()
plt.imshow( kolo & kwadrat )
plt.show()
plt.imshow( kolo | kwadrat )
plt.show()
plt.imshow( kolo ^ kwadrat )
plt.show()
```
